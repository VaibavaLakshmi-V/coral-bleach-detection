```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def analyze_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY)

    white_area = cv2.countNonZero(thresh)
    total_area = gray.shape[0] * gray.shape[1]
    white_ratio = white_area / total_area

    plt.imshow(image_rgb)
    plt.title(f"White Area Ratio: {white_ratio:.2f}")
    plt.axis('off')
    plt.show()

    if white_ratio > 0.3:
        print("High possibility of bleaching detected!")
    else:
        print("Coral appears healthy.")
```

```python
def analyze_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Lower range to capture not just pure white, but pale/bleached colors
    mask = cv2.inRange(hsv, (0, 0, 180), (180, 80, 255))  # Allowing a bit more saturation

    white_area = cv2.countNonZero(mask)
    total_area = image.shape[0] * image.shape[1]
    white_ratio = white_area / total_area

    plt.imshow(image_rgb)
    plt.title(f"Bleached Area Ratio: {white_ratio:.2f}")
    plt.axis('off')
    plt.show()

    if white_ratio > 0.15:
        print("Bleaching likely detected.")
    else:
        print("Coral appears healthy.")
```

```python
analyze_image("coral.jpg")
```

```python
analyze_image("healthy cora.jpg")
```
